**2.1 Решаем задачу подсчета слов в тексте**

В рамках данного задания вам нужно подсчитать кол-во слов в тексте Шекспира, по сути необходимо реализовать базовый функционал утилиты word count.

**Важно**, подсчитывайте число только тех слов, длина которых больше 4 символов. Проводить процесс удаления знаков препинания и прочих символов **не нужно**

Ниже вам предоставлены ячейки, в которые вы можете засипать структуру маппера/редьсюера, и ниже вызвать их в общем скрипта запуска map reduce таски

````python
%%writefile mapper.py
import sys

for line in sys.stdin:
    words = line.strip().split(" ")
    for word in words:
        if len(word.strip()) > 4:
            print(word, 1, sep='\t')
````

````python
%%writefile reducer.py
import sys

current_word = ""
word_count = 0

for line in sys.stdin:
    word, count = line.strip().split('\t')
    count = int(count)

    if current_word != word:
        if current_word:
            print(current_word, word_count, sep='\t')
        current_word = word
        word_count = 0

    word_count += count

if current_word:
    print(current_word, word_count, sep='\t')
````

В качестве проверки ваших питоновских скриптов до запуска MR таски можно произвести их запуск через консольные команды, по сути наша задача не будет выполняться через датаноты, а только на этой основной машине, но в случае некорректности своих скриптов вы легко увидете свои ошибки и сможете их поправить

````python
%%bash
## шаблон для запуска MR таски

# обязательная чистка директории, куда будем складывать результат отрабоки mr
hdfs dfs -rm -r /user/student_user27/word_count_task || true

# запус mr таски с указанием пути до нужного jar
hadoop jar /usr/lib/hadoop/tools/lib/hadoop-streaming-3.2.2.jar \
    -D mapreduce.job.name="word-count" \
    -files ./mapper.py,./reducer.py \
    -mapper "python3 mapper.py" \
    -reducer "python3 reducer.py" \
    -input /user/student_user27/input_data/shakespeare.txt \
    -output /user/student_user27/word_count_task
````

**2.2 Решаем задачу поиска максимального дохода у филиалов компаний по разным странам**

В данном задании нужно поработать с  информацией о доходах больших компаний и их филиалах в разных странных за некоторый период. Данные представлены в след. формате: `страна;компания;дата;доход`. Вам необходимо найти максимальный доход у каждой компании в отдельности по каждой представленной стране в данных.

Результат работы скрипта должен выглядеть следующим образом:

```bash
country  company                    income
Austria  Rodriguez, Sims and Arias  123456
USA      Rios-Lewis                 789998
Canada   Thompson-Mendez            148345
```

Также в рамках этой задачи вам могу пригодиться дополнительные параметры mr таски, отвечающие за настройку шаффла, и правил сортировки ключей внутри него. Почитать о примерах их использования можно [здесь](https://hadoop.apache.org/docs/current/hadoop-streaming/HadoopStreaming.html#More_Usage_Examples).

Также не рекомендуем использовать `\t` в качестве символа разделителя для сложного ключа (потому что по дефолту таб используется для разделения колонок данных, и ключом в таком случае будет только первая колонка до таба). Если вы будете собирать сложный ключ для нужной вам сортировки данных, лучше всего будет использовать другие симловы, к примеру `+, =`

Также как и в задании 2.1, вы можете создать набор ячеек, куда занесете код своих маппера и редьюсера, и также соберете код запуска MR таски. Нам нужно, чтобы вы предоставили нам в качестве решения этой задачи все эти составляющие в рамках этого ноутбука.

````python
%%writefile mapper.py
import sys

for line in sys.stdin:
    country, company, date, income = line.strip().split(';')
    print(f'{country}+{company}\t{income}')
````

````python
%%writefile reducer.py
import sys

current_country_company = ""
max_ = -1

for line in sys.stdin:
    country_company, income = line.strip().split('\t')
    income = int(income)

    if current_country_company != country_company:
        if current_country_company:
            country, company = current_country_company.split('+')
            print(country, company, max_, sep='\t')
        current_country_company = country_company
        max_ = -1

    max_ = max(max_, income)

if current_country_company:
    country, company = current_country_company.split('+')
    print(country, company, max_, sep='\t')
````

**2.3 Решаем задачу поиска самых посещаемых сайтов**

В данном задании нужно поработать с логом данных о посещении юзерами различных сайтов.
Формат данных: `url;временная метка`. Вам нужно вывести топ 5 сайтов по посещаемости в каждую из дат, которая представлена в наших данных.

Результат работы скрипта должен выглядеть следующим образом:

```bash
date        site                            count
2024-05-25  https://gonzales-bautista.com/  987
2024-05-25  https://smith.com/              654
2024-05-25  https://www.smith.com/          321
```

Также в рамках этой задачи вам могу пригодиться дополнительные параметры mr таски, отвечающие за настройку шаффла, и правил сортировки ключей внутри него. Почитать о примерах их использования можно [здесь](https://hadoop.apache.org/docs/current/hadoop-streaming/HadoopStreaming.html#More_Usage_Examples).

Также не рекомендуем использовать `\t` в качестве символа разделителя для сложного ключа (потому что по дефолту таб используется для разделения колонок данных, и ключом в таком случае будет только первая колонка до таба). Если вы будете собирать сложный ключ для нужной вам сортировки данных, лучше всего будет использовать другие симловы, к примеру `+, =`.

Возможно, у вас не получится решить данную задачу за одну mr таску, тогда вы просто описываете в решении скрипты ваших мапперов, редьюсеров под каждую из mr тасок, которые вам нужно запустить для получения нужного результата.

**Важно** помнить, что любой маппер и редьюсер должен работать за O(1) памяти, и если вы будете создавать какой-то список, куда будете складывать какие-то данные, то он не должен быть размера O(n). Если такой момент в вашем решении будет, пожалуйста, поясните его текстово, что с вашими переменными все ок, и у них нет размера O(n). 

````python
%%writefile mapper.py
import sys

for line in sys.stdin:
    try:
        url, datetime_str = line.strip().split(";")
        date = datetime_str.split()[0]
        print(f"{url}+{date}\t1")
    except Exception as e:
        print(f"Mapper error on line: {line.strip()} — {e}", file=sys.stderr)
````

````python
%%writefile reducer.py
import sys

current_url_date = ""
url_count = 0

for line in sys.stdin:
    url_date, count = line.strip().split('\t')
    count = int(count)

    if current_url_date != url_date:
        if current_url_date:
            print(url_count, current_url_date, sep='\t')
        current_url_date = url_date
        url_count = 0

    url_count += count

if current_url_date:
    print(url_count, current_url_date, sep='\t')
````

````python
%%bash

## Первая MR таска

hdfs dfs -rm -r /user/student_user27/site_visits/first_task || true

hadoop jar /usr/lib/hadoop/tools/lib/hadoop-streaming-3.2.2.jar \
    -D mapreduce.job.name="site-visits" \
    -files ./mapper.py,./reducer.py \
    -mapper "python3 mapper.py" \
    -reducer "python3 reducer.py" \
    -input /user/student_user27/input_data/site_visits.csv \
    -output /user/student_user27/site_visits/first_task
````

````python
%%writefile mapper.py
import sys

for line in sys.stdin:
    url_count, url_date = line.strip().split('\t')
    url, date = url_date.strip().split('+')
    print(date, url_count, url, sep='\t')
````

````python
%%writefile reducer.py
import sys

current_date = None
count_ = 0

for line in sys.stdin:
    date, url_count, url = line.strip().split('\t')
    if date != current_date:
        count_ = 0
        current_date = date
        
    if count_ <= 4:
        print(date, url, url_count)
        count_ += 1
````

````python
%%bash
hdfs dfs -rm -r /user/student_user27/site_visits/second_task || true

## вторая MR таска

hadoop jar /usr/lib/hadoop/tools/lib/hadoop-streaming-3.2.2.jar \
    -D mapreduce.job.name="site-visits-top5" \
    -D stream.num.map.output.key.fields=2 \
    -D mapreduce.job.output.key.comparator.class=org.apache.hadoop.mapreduce.lib.partition.KeyFieldBasedComparator \
    -D mapreduce.partition.keypartitioner.options=-k1,1 \
    -D mapreduce.partition.keycomparator.options="-k1,1 -k2,2nr" \
    -files ./mapper.py,./reducer.py \
    -mapper "python3 mapper.py" \
    -reducer "python3 reducer.py" \
    -input /user/student_user27/site_visits/first_task \
    -output /user/student_user27/site_visits/second_task
````
